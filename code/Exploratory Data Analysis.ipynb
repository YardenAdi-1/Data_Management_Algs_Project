# Exploratory Data Analysis

```python
import matplotlib.pyplot as plt
import pandas as pd
import warnings
import os

warnings.filterwarnings('ignore')
plt.style.use('ggplot')
```

```python
DATA_DIR = 'data/processed/'
raw_data_df = pd.read_csv('data/raw_data.csv')
age_thresholds =[20, 21, 23]
outcome_definitions = ['strict', 'inclusive']
```

## Distribution of the Treatment Variable

```python
fig, axs = plt.subplots(1, 4, figsize=(30, 5))

plt.subplots_adjust(wspace=0.3)

# Plot the distribution of the ages in the raw data
axs[0].hist(raw_data_df['Age at enrollment'], bins=30)
axs[0].set_title('Distribution of Age at Enrollment in the Raw Data')
axs[0].set_xlabel('Age at Enrollment')
axs[0].set_ylabel('Frequency')

# Plot the distribution of the treatment for every age threshold
for i, age_threshold in enumerate(age_thresholds):
    processed_data_df = pd.read_csv(os.path.join(DATA_DIR, f'processed_age_{age_threshold}_outcome_strict.csv'))
    axs[i + 1].hist(processed_data_df['Adult'], bins=3)
    axs[i + 1].set_title(f'Distribution of Treatment for Age Threshold {age_threshold}')
    axs[i + 1].set_xlabel('Adult')
    axs[i + 1].set_ylabel('Frequency')
    axs[i + 1].set_xticks([0, 1])
    axs[i + 1].set_xticklabels([f'Below {age_threshold}', f'{age_threshold} and above'])
```

## Distribution of the Outcome Variable

```python
fig, axs = plt.subplots(1, 3, figsize=(24, 5))

plt.subplots_adjust(wspace=0.3)

# Plot the distibution of the outcome variable in the raw data
axs[0].hist(raw_data_df['Target'], bins=5)
axs[0].set_title('Distribution of Outcome in the Raw Data')
axs[0].set_xlabel('Outcome')
axs[0].set_ylabel('Frequency')

# Plot the distribution of the outcome variable for every outcome definition
for i, outcome_definition in enumerate(outcome_definitions):
    processed_data_df = pd.read_csv(os.path.join(DATA_DIR, f'processed_age_20_outcome_{outcome_definition}.csv'))
    axs[i + 1].hist(processed_data_df['Target'], bins=3)
    axs[i + 1].set_title(f'Distribution of Outcome for Outcome Definition {outcome_definition}')
    axs[i + 1].set_xlabel('Outcome')
    axs[i + 1].set_ylabel('Frequency')
    axs[i + 1].set_xticks([0, 1])
    axs[i + 1].set_xticklabels(['Did not Graduate', 'Graduated'])
```
